```python
import nltk
import replacers
import pickle
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.svm import LinearSVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.pipeline import Pipeline
from sklearn.cross_validation import KFold
```

```python
# Declaring global variables
global corpus, labels, industry, category

# Change the following values to coose different industry and/or category
industry = "airline"
category = 7
```

```python
def tok(line):
    '''
    Tokenization function. Does correction, stemming and tokenizing.
    '''
    
    # Replacer to expand the contractions and stripping punctuations
    expand_replacer = replacers.RegexpReplacer()
    
    # Replacer to remove unnecessary repeated characters in words
    repeat_replacer = replacers.RepeatReplacer()
    
    # Replacer to correct the spelling of words
    spell_replacer = replacers.SpellingReplacer()

    line = expand_replacer.replace(line)
    words = [repeat_replacer.replace(word) for word in line.split()]
    words = [spell_replacer.replace(word) for word in words]

    # Using porter stemmer to stem words after splitting by spaces
    stemmer = nltk.stem.porter.PorterStemmer()
    words = [stemmer.stem(word) for word in words]

    return words
```

```python
def save_results():
    '''
    Function to save TF-IDF vectorization of the corpus of posts
    '''
    global industry, corpus
    
    # Taking the proper excel file from the industry name
    filename = "labeled data_AMT_" + industry + ".xls"
    xl = pd.ExcelFile(filename)
    df = xl.parse("Sheet 1")
    df = np.array(df)
    
    # Taking proper fields for posts and labels
    posts = df[:,5]

    # Converting corpus and labels to numpy arrays
    corpus = np.array(posts)

    # Doing TF-IDF vectorization of posts
    vect = TfidfVectorizer(ngram_range=(1,3), use_idf=True, analyzer='word', lowercase=True, encoding='utf-8', stop_words='english', tokenizer=tok, min_df=3)#, max_df=1995)
    corpus = vect.fit_transform(corpus)
    
    pick_name = "tfidf_" + industry
    
    f = open(pick_name, 'wb')
    pickle.dump(corpus, f)
    f.close()
```

```python
save_results()
```

```python
def get_results():
    '''
    Function to create TF-IDF vectorization of the corpus of posts
    '''
    global industry, corpus, labels
    
    # Taking the proper excel file from the industry name
    filename = "labeled data_AMT_" + industry + ".xls"
    xl = pd.ExcelFile(filename)
    df = xl.parse("Sheet 1")
    df = np.array(df)
    
    # Taking proper fields for posts and labels
    labels = df[:,category-1]

    # Converting corpus and labels to numpy arrays
    labels = np.array(labels, dtype=np.int32)

    # Binarizing labels
    labels = [0 if l<3 else 1 for l in labels]

    pick_name = "tfidf_" + industry
    
    f = open(pick_name)
    corpus = pickle.load(f)
    f.close()
```

```python
get_results()
```

```python
def validation():
    '''
    This function calculates the efficiency of the classification
    '''
    global corpus, labels, industry
    
    # Number of samples
    nm = len(labels)
    
    labels = np.array(labels)
    
    div = 0
    i = 0
    p = [0.0, 0.0, 0.0, 0.0]
    r = [0.0, 0.0, 0.0, 0.0]
    f = [0.0, 0.0, 0.0, 0.0]

    # Initializing the classifier and choosing chi2 to select k best features
    # Note: Every classifier is initialized using the default paramaters which might not give the best performance.
    #       Change these parameters by refering the documentation
    
    #Uncomment below line to enable SVM
    clf = Pipeline([('chi2', SelectKBest(chi2, k=150)), ('svm', LinearSVC())])
    
    #Uncomment below line to enable SVM
    #clf = Pipeline([('chi2', SelectKBest(chi2, k=700)), ('knn', KNeighborsClassifier())])
    
    #Uncomment below two lines to enable Decision Tree Classifier
    #corpus = corpus.toarray()
    #clf = Pipeline([('chi2', SelectKBest(chi2, k=700)), ('dtree', DecisionTreeClassifier())])
    
    #Uncomment below two lines to enable Naive Bayes Classifier
    #corpus = corpus.toarray()
    #clf = Pipeline([('chi2', SelectKBest(chi2, k=700)), ('nb', GaussianNB())])
    
    #Uncomment below two lines to enable Random Forest Classifier
    #corpus = corpus.toarray()
    #clf = Pipeline([('chi2', SelectKBest(chi2, k=100)), ('rf', RandomForestClassifier())])
    
    # Initializing k-fold cross validation with k=4
    kf = KFold(nm, n_folds=4)
    
    # Total true positives, false positives, false negatives and true negatives
    ttp = 0
    tfp = 0
    tfn = 0
    ttn = 0
    
    for train_index, test_index in kf:
        
        # Splitting corpus and labels into test and train set for one iteration of cross-validation
        X_train, X_test = corpus[train_index], corpus[test_index]
        Y_train, Y_test = labels[train_index], labels[test_index]
        
        # Training the classifier
        clf.fit(X_train, Y_train)
        pred = clf.predict(X_test)

        # Calculating true positives, false positives, false negatives and true negatives
        tp = 0
        fp = 0
        fn = 0
        tn = 0

        for tst, ref in zip(pred, Y_test):
            if tst == 1 and ref == 1:
                tp += 1
            elif tst == 1 and ref == 0:
                fp += 1
            elif tst == 0 and ref == 1:
                fn += 1
            else:
                tn += 1

        ttp+=tp
        tfp+=fp
        tfn+=fn
        ttn+=tn
                
        # Calculating precision, recall, f-measure
        p[i] = float(tp)/(tp+fp)
        r[i] = float(tp)/(tp+fn)
        f[i] = 2*p[i]*r[i]/(p[i]+r[i])

        i = i+1
        div = div+500

    print "Industry =", industry, ", Category =", category
    print "\nF-measure (macro) =", (f[0] + f[1] + f[2] + f[3])/4.0
    print "Precision (macro) =", (p[0] + p[1] + p[2] + p[3])/4.0
    print "Recall (macro) =", (r[0] + r[1] + r[2] + r[3])/4.0
    
    tpre = float(ttp)/(ttp+tfp)
    trec = float(ttp)/(ttp+tfn)
    
    print "\nF-measure (micro) =", 2*tpre*trec/(tpre + trec)
    print "Precision (micro) =", tpre
    print "Recall (micro) =", trec
```

```python
validation()
```
